```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression as LR
```

```python
data=pd.read_csv('training_prepared.csv')
```

```python
logistic_model=LR()
```

```python
data
```

```python
data.fillna(0,inplace=True)
data
```

```python
data['RainTomorrow']=data['RainTomorrow'].apply(lambda s: 1 if s=='Yes' else 0)
labels=data['RainTomorrow']
features=data.drop('RainTomorrow',axis=1)
features_arr=np.array(features)
print(np.where(np.isnan(features_arr)))
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test,train_labels,test_labels = train_test_split(features_arr, labels, test_size = 0.25, random_state = 30)
```

```python
logistic_model.fit(train,train_labels)
```

```python
pred=logistic_model.predict(test)
```

```python
from sklearn.metrics import accuracy_score, r2_score,log_loss
```

```python
accuracy_score(pred,test_labels)
```
